```python
import pandas as pd
from polygon import RESTClient
from datetime import date, datetime, timedelta
```

```python
df = pd.read_csv('sp500_sectors.csv')
tickers = df['symbol'].to_list()
```

```python
client = RESTClient(api_key='POLYGONIO API KEY')
```

```python
dateList = pd.read_csv('sp500_joined_closes.csv')['Date'].to_list()
df = pd.DataFrame(columns=['Date'], data=dateList)
for i in range(len(tickers)):
    try:
        print(i)
        bars = client.get_aggs(ticker=tickers[i], multiplier=1, timespan="day", from_=datetime(2010, 1, 1), to=datetime(2023, 3, 11))
        closeList = []
        for bar in bars:
            closeList.append(bar.close)
        if len(closeList) == 3319:
            df[tickers[i]] = closeList
            print(df)
    except:
        pass
```

```python
df.to_csv('sp500_sectors_prices.csv')
```
